Une **distribution de probabilité** $p(X = x)$ décrit la probabilité que $X$ prenne chacune de ses valeurs possibles $x$.

### **Un Mot sur la Notation...**

Nous allons bientôt voir beaucoup de nouvelles notations, alors prenons un peu plus de temps pour nous assurer que tout est clair avant de continuer. Si vous êtes nouveau dans ce sujet ou si vous en avez déjà vu une partie mais que vous êtes confus par la façon dont nous avons écrit les choses, alors lisez cette section.

**$X$ vs $x$**

* $X$ représente une variable aléatoire, c'est-à-dire une variable qui peut prendre plusieurs valeurs différentes en fonction du résultat d'un événement aléatoire.
* $x$ représente une valeur particulière que $X$ peut prendre. Vous pouvez penser à $x$ comme à une variable python `x`, qui à tout moment a une seule valeur.
* Une façon de décider d'écrire $X$ ou $x$ est de se poser la question : "La valeur de la variable est-elle connue ?" Si la réponse est oui, alors on écrit $x$. Si la réponse est non, alors on écrit $X$.

**$p(X = x)$ vs $p(X)$ vs $p(x)$**

* $p(X)$ est une abréviation qui signifie la même chose que $p(X = x)$, c'est-à-dire qu'elle représente la PMF/PDF d'une variable aléatoire $X$.
* $p(x)$ a une signification légèrement différente. Il indique la probabilité que la variable aléatoire $X$ prenne la valeur $x$. Au lieu de représenter la PDF, il représente la PDF évaluée en $x$. C'est une valeur de probabilité plutôt qu'une fonction.
* Vous pouvez également voir la notation $p_X(x)$, qui a la même signification que $p(x)$, mais inclut la variable aléatoire $X$ pour plus de clarté. Si nous écrivons $p_X(1)$, nous pouvons toujours voir que la variable aléatoire est $X$.


**$p$ vs $P$**

* Parfois, vous pouvez voir une notation qui utilise à la fois $p$ et $P$. Dans ce cas, $P$ et $p$ sont utilisés pour différencier les distributions pour les variables aléatoires *discrètes* et *continues*, respectivement.
* Vous pouvez également voir d'autres notations telles que $\text{Pr}$ ou $\mathbb{P}$. Celles-ci n'ont pas de signification particulière sauf indication contraire dans le texte où elles apparaissent.

### **La Distribution de Bernoulli**

Une distribution de Bernoulli est utilisée pour décrire des événements aléatoires avec deux résultats possibles, par exemple lorsque nous avons une variable aléatoire $X$ qui prend l'une des deux valeurs $x \in \{0, 1\}$ :
\begin{align}
p(X = 1) &= p \\
p(X = 0) &= 1 - p
\end{align}
où $0 \le p \le 1$. Ici, $p$ est un paramètre qui contrôle la probabilité de chaque résultat. Par exemple, dans le cas d'un lancer de pièce *équitable*, $p = 0,5$ de sorte que les deux résultats ont 50 % de chances de se produire. Comme nous le verrons bientôt, la plupart des distributions ont plusieurs paramètres.

Une distribution de probabilité peut être définie de manière unique par sa fonction de masse de probabilité (PMF), ou fonction de densité de probabilité (PDF) dans le cas des distributions continues. Pour la distribution de Bernoulli, nous écrivons la PMF comme suit :
\begin{align}
p(X = x) = \text{Bernoulli}(X\mid p) &= \begin{cases}
p\, & \text{si } x = 1 \\
1 - p\, & \text{si } x = 0
\end{cases} \\
&= p^x(1-p)^{1-x}.
\end{align}

Maintenant, construisons une distribution de Bernoulli en code afin de pouvoir jouer avec et acquérir une certaine intuition.

**Note:** Dans ce TP, nous allons utiliser `numpyro` pour construire nos distributions. Cependant, il existe plusieurs autres packages `jax` qui fonctionnent de manière similaire (par exemple, `distrax`) ainsi que plusieurs options pour `tensorflow` (par exemple, `tensorflow_probability`) et `pytorch` (par exemple, `torch.distribution`). Ne vous inquiétez pas trop des détails du fonctionnement de `numpyro`, par exemple les noms des distributions et leurs arguments. Essayez plutôt de comprendre ce que fait le code.

```python
p = jnp.array(0.5)
bernoulli = dist.Bernoulli(probs=p)
```

Maintenant construisons une distribution que nous pouvons échantilloner:

```python
sample = bernoulli.sample(key=rng)
print(sample)
```

Et nous pouvons évaluer la probabilité d'observer un échantillon.

**Note:** les objets de distribution dans `numpyro` (et en effet la plupart des autres bibliothèques pour les distributions de probabilité) renvoient des *log*-probabilités plutôt que des probabilités brutes. Cela signifie que nous devons prendre l'exponentielle si nous voulons connaître la probabilité.

```python
log_prob = bernoulli.log_prob(sample)
print(f"log p(X = {sample}) = {log_prob}")
print(f"p(X = {sample}) = {jnp.exp(log_prob)}")
```

Comme anticipé, nous obtenons une probabilité de 0.5.

Nous pouvons facilement obtenir de **multiple échantillons**:

```python
# { run: "auto" }
n = 10 #@param {type:"slider", min:1, max:10, step:1}
samples = bernoulli.sample(key=rng, sample_shape=(n,))
print(samples)
```

*Mais que se passerait-il si nous voulions évaluer la probabilité d'observer tous nos échantillons ?*

L'objet `bernoulli` que nous avons créé précédemment traite chaque échantillon individuellement et renvoie les probabilités d'observer chaque échantillon séparément:

```python
individual_sample_probs = jnp.exp(bernoulli.log_prob(samples))
print(individual_sample_probs)
```

Mais nous pouvons utiliser l'une des lois de la probabilité pour calculer la probabilité d'observer tous les échantillons ensemble :
$$
p(X_1=x_1, X_2=x_2, \dots, X_N=x_n) = \prod_{n=1}^N p(X_n=x_n).
$$
C'est ce qu'on appelle la *règle du produit de la probabilité*, et elle dit que pour des variables aléatoires *indépendantes*, la probabilité jointe (c'est-à-dire la probabilité de les observer toutes ensemble) est égale au produit des probabilités individuelles.

Maintenant, calculons la probabilité jointe de nos échantillons.

```python
joint_prob = jnp.prod(individual_sample_probs)
print(joint_prob)
```

La réponse, comme c'est souvent le cas, est la stabilité numérique ! Lorsque l'on travaille avec des nombres entre 0 et 1 (c'est-à-dire des probabilités), il est facile de rencontrer un *dépassement négatif*. Un dépassement négatif se produit lorsqu'un nombre très proche de zéro est arrondi pour devenir *exactement* zéro car la quantité de mémoire utilisée pour le représenter n'est pas suffisante pour maintenir la précision.

Par exemple, considérons le calcul de la probabilité de plusieurs événements indépendants $A$, $B$, $C$, $\ldots$:
$$
p(A, B, C, \ldots) = p(A) \times p(B) \times p(C) \times \ldots,
$$
car nous multiplions des nombres compris entre 0 et 1, nous savons que le résultat sera plus petit pour chaque multiplication. Au fur et à mesure que nous effectuons de plus en plus de multiplications, il devient plus probable que le calcul déborde et produise par erreur un résultat de 0.

Nous pouvons éviter ce problème en travaillant plutôt avec des log-probabilités. En poursuivant l'exemple ci-dessus, nous obtenons
$$
\log p(A, B, C, \ldots) = \log p(A) + \log p(B) + \log p(C) + \ldots,
$$
où, par les lois des logarithmes, nous avons remplacé toutes les multiplications par des additions, évitant ainsi tout problème numérique. En effet, les nombres très proches de zéro deviennent de grands nombres négatifs lorsque vous appliquez la fonction logarithmique. Par exemple, $\log_{10}(0.0000001) = -16.1181$. L'addition de ces plus grandes log-probabilités (= multiplication des probabilités) préservera la précision plus longtemps, évitant ainsi le risque d'arrondir accidentellement les probabilités correspondantes à zéro.

### **Distribution catégorielle**

Une distribution catégorielle est utilisée pour modéliser des événements aléatoires avec plusieurs résultats *discrets*, comme l'événement de lancer de dé ci-dessus. Par discret, nous entendons qu'il y a un nombre fini de résultats. Nous aborderons bientôt les résultats continus – par exemple, une valeur comme la température qui est modélisée par des nombres réels.

Comme pour la distribution de Bernoulli, nous pouvons caractériser la distribution catégorielle par sa PMF :
\begin{align}
p(X = x) = \text{categorical}(X\mid\mathbf{p}) = \prod_{k=1}^K p_k^{[x=k]},
\end{align}
où $K$ est le nombre de résultats, $\textbf{p} = [p_1, p_2, \dots, p_K]$, $p_k$ est la probabilité du $k$-ième résultat, $\sum_k p_k = 1$, et $[x = k]$ est appelé un crochet d'Iverson qui vaut 1 si $x = k$ et 0 sinon.

**Essayez ceci:** Convainquez-vous qu'une distribution catégorielle avec $K = 2$ est équivalente à une distribution de Bernoulli.

Maintenant, commençons à jouer avec cette distribution en code !

```python
ps = jnp.array([0.1, 0.2, 0.3, 0.4])
categorical = dist.Categorical(probs=ps)
```

Comme précédemment nous pouvons prendre des échantillons:

```python
samples = categorical.sample(key=rng, sample_shape=(10,))
print(samples)
```

Et calculer la probabilité de chacun de ces événements aléatoires:

```python
print(f"p(X=0) = {jnp.exp(categorical.log_prob(0)):.1f}")
print(f"p(X=1) = {jnp.exp(categorical.log_prob(1)):.1f}")
print(f"p(X=2) = {jnp.exp(categorical.log_prob(2)):.1f}")
print(f"p(X=3) = {jnp.exp(categorical.log_prob(3)):.1f}")
```

### 🔧 **Tâche de Groupe (A3)**

Jouez avec les curseurs suivants, qui contrôlent les paramètres de la distribution catégorielle. Notez que parce que $\sum_k p_k = 1$, nous utilisons une autre façon de représenter la distribution catégorielle. Au lieu de spécifier les probabilités $p_k$, nous spécifions les logits $l_k$. Chaque $p_k$ est alors calculé comme suit :
$$
p_k = \frac{\exp(l_k)}{\sum_{k'}\exp(l_k)},
$$

### **Expected Values**

La valeur attendue d'une variable aléatoire est la valeur que nous nous attendons à obtenir en moyenne lorsque nous échantillonnons un grand nombre de résultats différents de l'événement aléatoire.

Par exemple, si nous considérons notre exemple de lancer de pièce ci-dessus, où $X = 1$ si la pièce tombe sur pile et $X = 0$ si la pièce tombe sur face, alors intuitivement la valeur attendue est $0,5$. Pour voir cela, imaginez un jeu où vous obtenez 1 pièce pour pile et aucune pièce pour face. Si vous jouez mille parties, vous devriez vous attendre à obtenir environ $500$ pièces au total. Ainsi, votre profit attendu pour un seul match est de $500/1000 = 0,5$ pièces.

Plus précisément, nous pouvons écrire la valeur attendue comme suit :
$$
\mathbb{E}[X] = \sum_x x \cdot p(X=x),
$$
pour une variable aléatoire discrète $X$ ou
$$
\mathbb{E}[X] = \int x \cdot p(X=x) \mathrm{d}x,
$$
pour une variable aléatoire continue. À partir de cette définition, nous pouvons considérer la valeur attendue comme une moyenne pondérée de toutes les valeurs de $X$, où la pondération pour chaque valeur est la probabilité de l'observer.

**$E$ vs $\mathrm{E}$ vs $\mathbb{E}$**

* Vous avez peut-être aussi vu une variété de notations pour les valeurs attendues ($E$, $\mathrm{E}$, $\mathbb{E}$) et les variances ($V$, $\mathrm{V}$, $\mathbb{V}$, $\text{Var}$). Ne vous inquiétez pas, tous ont exactement la même signification.

Nous pouvons également calculer les valeurs attendues de fonctions qui prennent des variables aléatoires comme entrées. Par exemple, la variance d'une variable aléatoire peut être calculée en utilisant la valeur attendue suivante :
$$
\mathbb{V}[X] = \mathbb{E}[(X - \mu)^2],
$$
où $\mu = \mathbb{E}[X]$ est la moyenne, comme décrit ci-dessus.

### **Calcul des Valeurs Attentes : Estimation Monte Carlo**

Au lieu de calculer une valeur attendue exactement, nous pouvons l'estimer en utilisant une approximation de *Monte Carlo*. Nous pouvons décrire cette procédure en utilisant la notation mathématique suivante :
$$
\mathbb{E}[X] \approx \frac{1}{N} \sum_{n=1}^{N} x_n\,, \quad x_n \sim p(X).
$$
C'est-à-dire que nous échantillonnons $N$ valeurs $x_n$ à partir de $p(X)$ puis prenons la moyenne. Parce que les valeurs apparaissent proportionnellement à leurs probabilités, plus $N$ est grand, plus l'estimation sera précise.

Nous pouvons également implémenter cela en code et calculer la valeur attendue de la distribution normale $\mathcal{N}(x \mid \mu, \sigma)$.

### **La Distribution Bêta**

La dernière distribution de notre *grand tour des distributions de probabilité (continues)* (pour l'instant) est la distribution bêta. Nous la regardons pour deux raisons.

Premièrement, nous voulons simplement jeter un coup d'œil à une distribution de probabilité continue qui n'est *pas* normale afin d'avoir une nouvelle perspective. Deuxièmement, cette distribution sera utile plus tard dans le TP – donc si vous ne la connaissez pas déjà, restez dans cette section.

La distribution bêta est utile pour décrire une variable aléatoire continue entre 0 et 1, comme les pourcentages et les ratios.

Ne vous inquiétez pas de la forme exacte de la fonction bêta, tout ce que vous devez savoir est que nous pouvons facilement la calculer en utilisant `numpy` ou des bibliothèques similaires, et qu'elle garantit simplement que la distribution de probabilité est normalisée – c'est-à-dire que la probabilité totale de tous les événements s'additionne à 1. ### **La Distribution Normale Multivariée**
La distribution normale multivariée généralise la distribution normale pour nous permettre de considérer plusieurs variables aléatoires, où les résultats de ces multiples variables aléatoires prennent une *distribution normale conjointe*.

En d'autres termes, nous pouvons maintenant modéliser des variables qui ne sont pas indépendantes : si nous connaissons la valeur d'une variable, cela nous renseigne sur les autres variables !

Plus concrètement, la distribution normale multivariée nous permet de considérer plusieurs variables aléatoires telles que lorsque nous conditionnons certaines de ces variables, les variables restantes ont une distribution normale (ne vous inquiétez pas si vous ne savez pas encore ce que signifie "conditionner" ; nous en discuterons bientôt).

Ces variables sont distribuées dans une sorte de boule floue étirée dans un espace de dimension supérieure.

### **Relations entre Variables**
En règle générale, plus une variable nous renseigne sur une autre, plus la *covariance* ou la *corrélation* entre les deux est grande. Assez de texte, jetons un coup d'œil aux équations, qui donneront, espérons-le, un aperçu plus approfondi. Tout d'abord, regardons la PDF générale pour une variable aléatoire $\mathbf{X}$ de dimension $D$ :
$$
p(\mathbf{X} = \mathbf{x}) = \mathcal{N}(\mathbf{x}\mid \boldsymbol{\mu}, \Sigma) = \frac{1}{\sqrt{(2\pi)^D|\Sigma|}}\exp\left(-\frac{1}{2}(\mathbf{x} - \boldsymbol{\mu})^\intercal\Sigma^{-1}(\mathbf{x} - \boldsymbol{\mu})\right),
$$
où $\mathbf{x}$ et $\boldsymbol{\mu}$ sont maintenant des vecteurs de nombres plutôt que des nombres simples, $\Sigma$ est une *matrice de covariance* qui remplace $\sigma$ de notre définition univariée ci-dessus, et $|\Sigma|$ est son déterminant. La matrice de covariance ressemble à ceci :
$$
\Sigma = \left[\matrix{
    \sigma_1^2 & \rho_{12} \sigma_1 \sigma_2 & \cdots &  \rho_{1D} \sigma_1 \sigma_D  \\
    \rho_{21} \sigma_2\sigma_1 & \sigma_2^2 & \cdots &  \rho_{2D} \sigma_2 \sigma_D \\
    \vdots & \vdots & \ddots & \vdots \\
    \rho_{D1} \sigma_D \sigma_1 & \rho_{D2} \sigma_D \sigma_2 & \cdots & \sigma_D^2}\right],
$$
où $\sigma_i^2$ est la variance pour la $i$-ème dimension, et $\rho_{ij} = \rho_{ji}$ est la *corrélation* entre la $i$-ème et la $j$-ème dimensions. La matrice de covariance nous indique comment la "boule" de variables aléatoires est étirée et tournée dans l'espace.\\

### **Les Règles de la Somme et du Produit**

Maintenant que nous avons introduit une distribution multivariée, nous pouvons l'utiliser pour explorer deux concepts importants lorsque l'on travaille avec des distributions de probabilité : les distributions marginales et conditionnelles. Cela nous conduira à deux règles de probabilité importantes dont nous aurons besoin plus tard dans le TP (et dans la vie) : les règles de la somme et du produit.

Rappelez-vous que formellement, les **probabilités conjointes** nous indiquent la probabilité que deux événements ou plus se produisent pour des variables aléatoires. Pour deux variables, nous écrivons $P(A=a,B=b)$. Ce n'est pas nouveau - vous venez de voir des probabilités conjointes en action pour la distribution normale multivariée ! 🤔

#### **Distributions Marginales & la Règle de la Somme**


Considérons $p(X_1, X_2)$ une distribution *jointe* sur deux variables aléatoires $X_1$ et $X_2$. Maintenant, que se passerait-il si nous ne nous soucions que d'une des variables aléatoires, par exemple $X_1$ ? Nous pouvons utiliser la règle de la *somme* pour *marginaliser* $X_2$ à partir de la distribution jointe afin d'obtenir la distribution *marginale* $p(X_1)$. Concrètement, pour un $X_2$ discret, la règle de la somme peut s'écrire comme suit :
$$
p(X_1 = x_1) = \sum_{x_2} p(X_1 = x_1, X_2 = x_2).
$$
De même, pour un $X_2$ continu, nous aurions
$$
p(X_1 = x_1) = \int p(X_1 = x_1, X_2 = x_2)\, \mathrm{d}x_2.
$$
Cette règle nous dit que si nous voulons une distribution marginale, il suffit d'énumérer toutes les valeurs des variables que nous ne voulons pas conserver dans notre distribution marginale et de sommer les distributions résultantes.

Que se passe-t-il si nous avons trois (ou plus) variables aléatoires ? Rien de spécial, quelles que soient les variables que nous ne voulons pas conserver dans notre distribution marginale, nous sommons, par exemple
$$
p(X_1 = x_1, X_3 = x_3) = \sum_{x_2} \sum_{x_4} p(X_1 = x_1, X_2 = x_2, X_3 = x_3, X_4 = x_4).
$$

Pour une distribution normale multivariée, la règle de la somme a une forme très simple et claire, ce qui devrait rendre les choses plus claires. Plus précisément, si
$$p(X_1, X_2) = \mathcal{N}\left(\left[\matrix{x_1 \\ x_2}\right]\middle| \left[\matrix{\mu_1 \\ \mu_2}\right], \left[\matrix{\sigma_1^2 & \rho \sigma_1 \sigma_2 \\ \rho \sigma_1 \sigma_2  &\sigma_2^2}\right]\right),$$
comme défini ci-dessus, alors
$$
p(X_1) = \mathcal{N}\left(x_1 \mid \mu_1, \sigma_1\right).
$$
En d'autres termes, nous "laissons tomber" simplement les variables qui ne nous intéressent pas, et obtenons une autre distribution normale avec seulement celles qui nous intéressent !

*Remarque :* certains lecteurs perspicaces ont peut-être remarqué quelque chose de légèrement étrange... dans la formule de distribution jointe, tous les $σ$ sur la diagonale sont au carré, alors que dans la formule marginale, nous écrivons simplement $\sigma$ sans le carré. C'est simplement une convention, vous pouvez également écrire $\mathcal{N}\left(x_1 \mid \mu_1, \sigma_1^2\right)$, la signification est la même.

Une autre façon d'exprimer les marginales est que lorsque nous sommons ou intégrons toutes les possibilités pour une variable dans une probabilité jointe, nous pouvons obtenir la distribution de probabilité de l'autre ! Vous pouvez deviner le nom en pensant à intégrer ou à sommer vers l'axe de la variable $A$ ou $B$.

#### **Distributions Conditionnelles & la Règle du Produit**

Comme nous l'avons mentionné ci-dessus, nous utilisons des distributions *jointes* lorsqu'une variable aléatoire nous renseigne sur une autre variable aléatoire. Mais comment utilisons-nous réellement cette information ? Par exemple, si $X_2 = 2$, comment cela change-t-il la distribution sur $X_1$ ? C'est exactement ce que nous disent les distributions *conditionnelles*. Pour être concret, la distribution *conditionnelle* de $X_1$ sachant $X_2$ peut être calculée comme suit :
$$
p(X_1 = x_1|X_2 = x_2) = \frac{p(X_1 = x_1, X_2 = x_2)}{p(X_2 = x_2)},
$$
en d'autres termes, c'est le rapport de la distribution *jointe* et de la distribution *marginale* de la variable de conditionnement. En réarrangeant la formule ci-dessus, nous pouvons maintenant écrire une formule pour une distribution *jointe* dans le cas où les variables *ne sont pas* indépendantes
$$
p(X_1 = x_1, X_2 = x_2) = p(X_1 = x_1|X_2 = x_2)p(X_2 = x_2).
$$
Quel $X_1$ et $X_2$ *sont* indépendants? Alors
$$
p(X_1 = x_1|X_2 = x_2) = p(X_1 = x_1),
$$
et nous pouvons écrire la jointure comme:
$$
p(X_1 = x_1, X_2 = x_2) = p(X_1 = x_1)p(X_2 = x_2),
$$
ce qui permet de retrouver la formule que nous avons utilisé dans la tâche de groupe 3!

Comme ci-dessus, pour plus de 2 variables, les choses restent simples
$$
p(X_1 = x_1, X_3 = x_3|X_2 = x_2, X_4 = x_4) = \frac{p(X_1 = x_1, X_2 = x_2, X_3 = x_3, X_4 = x_4)}{p(X_2 = x_2, X_4 = x_4)}
$$
ou de manière équivalente
$$
p(X_1 = x_1, X_2 = x_2, X_3 = x_3, X_4 = x_4) = p(X_1 = x_1, X_3 = x_3|X_2 = x_2, X_4 = x_4)p(X_2 = x_2, X_4 = x_4).
$$

Et comme ci-dessus, pour la normale à deux dimensions, nous obtenons des équations simples qui peuvent nous aider à comprendre ce qui se passe lorsque nous travaillons avec des distributions conditionnelles.
$$
p(X_1=x_1|X_2=x_2) = \mathcal{N}\left(x_1\mid \mu_1 + \frac{\sigma_1}{\sigma_2}\rho(x_2 - \mu_2), (1 - \rho^2)\sigma_1^2\right).
$$

**Relier la probabilité jointe, marginale et conditionnelle:**

Ces quantités sont liées par : $p(A,B) = p(A \mid B) p(B) = p(B \mid A) p(A)$.

**Un mot sur la notation:**

**$p(A \cap B)$ vs $p(A , B)$**

* Ce sont des notations équivalentes pour la distribution jointe. C'est-à-dire la probabilité que l'événement $A$ se produise et que l'événement $B$ se produise.

**Qu'est-ce que le '$\mid$' dans $p(X\mid a, b, c)$ ?**

* La barre que l'on voit souvent au milieu de la notation de la distribution de probabilité est utilisée pour séparer les variables aléatoires de la distribution des variables dont les variables aléatoires sont conditionnées.
* En d'autres termes, nous avons une fonction avec deux types d'entrées différents. Les entrées à droite de $|$ sont des variables de conditionnement. Les entrées à gauche de $|$ sont les valeurs des variables aléatoires sous la distribution de probabilité. La sortie de la fonction est la probabilité des valeurs de ces variables aléatoires.
* On lit $p(A \mid B, C)$ comme 'la probabilité de A *sachant* B et C'.

## 🤖 **`Numpyro`: Batch and Event Shapes**

Prenons un moment pour jouer avec du code pour la normale multivariée et son amie plus simple, la normale univariée. Nous allons profiter de cette occasion pour apprendre quelque chose sur la façon dont `numpyro` (et la plupart des autres packages pour travailler avec des distributions de probabilité) gèrent les lots de données – quelque chose que nous devrons comprendre pour entraîner des modèles d'apprentissage automatique !

Toutes les distributions dans `numpyro` ont une `event_shape` – qui décrit comment mes dimensions la variable aléatoire est, par exemple, pour une distribution normale bidimensionnelle, cela serait 2 – et une `batch_shape` qui décrit combien de jeux de paramètres la distribution a – il est probablement plus facile de montrer ce que cela signifie avec les exemples suivants plutôt que de le dire.

Regardons d'abord une normale univariée simple $\mathcal{N}(x|0, 1)$, comme nous l'avons vu précédemment dans ce TP. Nous allons évaluer les probabilités de $X = 1$ et $X = 2$:

```python
values = jnp.array([1., 2.])
normal = dist.Normal(0., 1.)
print(f"event_shape = {normal.event_shape}")
print(f"batch_shape = {normal.batch_shape}")
print(f"p(X = {values}) = {jnp.exp(normal.log_prob(values))}")
```

Nous voyons que cette distribution a une forme d'événement vide, que vous pouvez considérer comme étant la même qu'une forme d'événement de 1 (comme un scalaire est le même qu'un vecteur de longueur 1). La forme du lot est également vide, puisque nous n'avons spécifié qu'un seul ensemble de paramètres ($\mu = 0, \sigma = 1$).

Maintenant, puisque nous avons essayé d'évaluer la probabilité de deux valeurs à la fois, et que ni la forme de l'événement ni la forme du lot ne sont égales à 2, cela équivaut à appeler `dist.log_prob(1.)` et `dist.log_prob(2.)` séparément. `numpyro` nous facilite la vie en *diffusant* le calcul `log_prob` pour faire à la fois $p(X=1) = \mathcal{N}(X=1|0, 1)$ et $p(X=2) = mathcal{N}(X=2|0, 1)$ en même temps.

Nous pourrions également spécifier un *lot* de deux ensembles de paramètres afin de travailler essentiellement avec $\mathcal{N}(x|0, 1)$ et $\mathcal{N}(x|1, 2)$ en même temps :

```python
batch_normal = dist.Normal(jnp.array([0., 1.]), jnp.array([1., 2.]))
print(f"event_shape = {batch_normal.event_shape}")
print(f"batch_shape = {batch_normal.batch_shape}")
print(f"[p(X_1 = {values[0]}), p(X_2 = {values[1]})] = {jnp.exp(batch_normal.log_prob(values))}")
print(f"p(X_1 = {values[0]}, X_2 = {values[1]}) = {jnp.prod(jnp.exp(batch_normal.log_prob(values)))}")
```

Maintenant, remarquez que bien que la forme de l'événement soit vide (comme prévu puisque nous travaillons toujours avec une normale univariée), la taille du lot est de 2 !

En conséquence, le calcul que nous faisons équivaut à calculer séparément $p(X_1=1) = \mathcal{N}(X_1=1|0, 1)$ et $p(X_2=2) = \mathcal{N}(X_2=2|1, 2)$ ! Encore une fois, c'est juste `numpyro` qui nous facilite la vie.

Si nous voulons calculer $p(X_1=1,X_2=2)$, c'est-à-dire la probabilité jointe que $X_1 = 1$ et $X_2 = 2$, alors nous devons soit multiplier manuellement les probabilités (en supposant que $X_1$ et $X_2$ sont indépendants) – comme nous l'avons fait dans la cellule ci-dessus – soit travailler avec une normale multivariée.

Regardons d'abord le cas où les $X_1$ et $X_2$ sont indépendants :

```python
multivariate_diagonal_normal = dist.MultivariateNormal(jnp.array([0., 1.]), jnp.array([[1., 0.], [0., 2.**2]]))
print(f"event_shape = {multivariate_diagonal_normal.event_shape}")
print(f"batch_shape = {multivariate_diagonal_normal.batch_shape}")
print(f"p(X_1 = {values[0]}, X_2 = {values[1]})) = {jnp.exp(multivariate_diagonal_normal.log_prob(values))}")
```

Comme prévu, pour cette normale à deux dimensions, nous avons maintenant une taille d'événement de 2 ! Et parce que nous n'avons spécifié qu'un seul ensemble de paramètres, la forme du lot est vide.

En conséquence de cette forme d'événement, lorsque nous demandons la `log_prob` de nos deux nombres, nous obtenons automatiquement une seule probabilité *jointe*, qui correspond à la valeur que nous avons calculée manuellement ci-dessus.

En utilisant une normale multivariée, nous pouvons également introduire une certaine covariance entre $X_1$ et $X_2$:

```python
multivariate_full_normal = dist.MultivariateNormal(jnp.array([0., 1.]), jnp.array([[1., 1.], [1., 2.**2]]))
print(f"event_shape = {multivariate_full_normal.event_shape}")
print(f"batch_shape = {multivariate_full_normal.batch_shape}")
print(f"p(X_1 = {values[0]}, X_2 = {values[1]}) = {jnp.exp(multivariate_full_normal.log_prob(values))}")
```

Cela entraîne une valeur légèrement différente pour la probabilité jointe, mais tout le reste semble identique.
